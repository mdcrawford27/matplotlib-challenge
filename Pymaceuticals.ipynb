```python
# dependencies and setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# create a path for the data files
mouse_metadata_path = "resources/Mouse_metadata.csv"
study_results_path = "resources/Study_results.csv"

# read the csv data files
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# combine the data files into one based on mouse ID
combined_data = pd.merge(mouse_metadata, study_results, on = "Mouse ID")

# display the data table
combined_data
```

```python
# checking the number of mice
mice_count = len(combined_data["Mouse ID"].unique())
mice_count
```

```python
# getting the duplicate mice that show up for Mouse ID and Timepoint
duplicate_mice = combined_data.loc[combined_data.duplicated(subset=["Mouse ID","Timepoint"]),"Mouse ID"].unique()
duplicate_mice
```

```python
# create a clean dataframe by dropping the duplicate mouse by its ID
clean_data = combined_data[combined_data["Mouse ID"] != "g989"]
```

```python
clean_data
```

```python
# checking the number of mice in the clean dataframe
mice_count2 = len(clean_data["Mouse ID"].unique())
mice_count2
```

```python
# SUMMARY STATISTICS

# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.

mean = clean_data.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]
median = clean_data.groupby("Drug Regimen").median()["Tumor Volume (mm3)"]
variance = clean_data.groupby("Drug Regimen").var()["Tumor Volume (mm3)"]
stdev = clean_data.groupby("Drug Regimen").std()["Tumor Volume (mm3)"]
sem = clean_data.groupby("Drug Regimen").sem()["Tumor Volume (mm3)"]

summary_table1 = pd.DataFrame({"Mean Tumor Volume":mean,
                             "Median Tumor Volume":median,
                             "Tumor Volume Variance":variance,
                             "Tumor Volume Std. Deviation":stdev,
                             "Tumor Volume Std. Error":sem})
summary_table1
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
summary_table2 = clean_data.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_table2
```

```python
# QUARTILES, OUTLIERS AND BOXPLOTS

# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
end_tumor = clean_data.groupby(["Mouse ID"]).max()["Timepoint"]
end_tumor = end_tumor.reset_index()
end_tumor_data = pd.merge(end_tumor,clean_data,on=["Mouse ID","Timepoint"],how="inner")
end_tumor_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
plotted_tumors = []

 # Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    
     # Locate the rows which contain mice on each drug and get the tumor volumes
    add_data = end_tumor_data.loc[end_tumor_data["Drug Regimen"] == drug, "Tumor Volume (mm3)"]
    
    # add this data to the list for plotting
    plotted_tumors.append(add_data)
    
    # Determine outliers using upper and lower bounds
    quartiles = add_data.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = add_data.loc[(add_data < lower_bound) | (add_data > upper_bound)]
    print(f"{drug}'s potential outliers are {outliers}")
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(plotted_tumors, labels = treatments)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

```python
# LINE AND SCATTER PLOTS
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
capomulin = clean_data.loc[clean_data["Drug Regimen"] == "Capomulin"]
capomulin
```

```python
mouse = capomulin.loc[capomulin["Mouse ID"] == 's185']
plt.plot(mouse["Timepoint"],mouse["Tumor Volume (mm3)"])
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment in Mouse S185")
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
capomulin_tumor_avg = capomulin.groupby(["Mouse ID"]).mean()
plt.scatter(capomulin_tumor_avg["Weight (g)"], capomulin_tumor_avg["Tumor Volume (mm3)"])
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

```python
# CORRELATION AND REGRESSION

 # Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation = round(st.pearsonr(capomulin_tumor_avg["Weight (g)"],capomulin_tumor_avg["Tumor Volume (mm3)"])[0],2)
print(f"The correlation between mouse weight and average tumor volume for the Capomulin regimen is {correlation}")


lrm = st.linregress(capomulin_tumor_avg["Weight (g)"],capomulin_tumor_avg["Tumor Volume (mm3)"])
y_values = capomulin_tumor_avg["Weight (g)"]*lrm[0]+lrm[1]
plt.scatter(capomulin_tumor_avg["Weight (g)"],capomulin_tumor_avg["Tumor Volume (mm3)"])
plt.plot(capomulin_tumor_avg['Weight (g)'],y_values,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
